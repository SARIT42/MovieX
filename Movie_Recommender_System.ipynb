```python
import numpy as np 
import pandas as pd
```

```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.shape
```

```python
# We want to create tags for recommender systems, columns required for the tags :
# genres, id, keywords, title, overview, production_companies, cast, crew


movies=movies[['movie_id','title','overview','production_companies','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# Given: [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
# Needed: ['Action', 'Adventure' , 'Fantasy' , 'Science-Fiction']
import ast
def convert(obj): #getting the required format of the columns
  l=[]
  for i in ast.literal_eval(obj): # converts the string dictionary list to dictionary list
    l.append(i['name'])
  return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convertcc(obj):  #getting the top 5 casts of each movie.
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=5:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast'] = movies['cast'].apply(convertcc)
```

```python
def fetch_dir(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew'] =  movies['crew'].apply(fetch_dir)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
def convertprod(obj):  
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=1:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['production_companies']= movies['production_companies'].apply(convertprod)
```

```python
movies.head()
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies']= movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['production_companies'] + movies['cast'] + movies['crew']
```

```python
df= movies[['movie_id','title','tags']]
df.head()
```

```python
df['tags'] =  df['tags'].apply(lambda x:" ".join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
df.tags[0]
```

```python
df.head()
```

```python
#stemming
import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]

  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)

```

```python
df['tags']=df['tags'].apply(stem)
```

```python
df.head()
```

Final Data prepared


```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[1])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
  movie_index = df[df['title']== movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Halloween')
```

```python
recommend('The Dark Knight Rises')
```
